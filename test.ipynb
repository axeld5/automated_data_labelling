```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from datasets import load_dataset

dataset = load_dataset("PresageLabs/Predibench")
df = dataset["train"].to_pandas()
df.to_csv("predibench.csv", index=False)
```

```python
with open("prompt.txt", "r") as f:
    prompt = f.read()

filename = "predibench"
# Load the good_dataset_card.txt as an example dataset card
with open("good_dataset_card.txt", "r") as f:
    example_card = f.read()

full_prompt = prompt + "\n\n" + "Here is an example of a good dataset card: \n\n" + example_card + "\n\n" + "Now, write the card for the following dataset which file can be found at {}.csv".format(filename) + "\n\n" + f"Right before writing your final answer, write the card in {filename}.txt"
```

```python
from smolagents import CodeAgent, LiteLLMModel
import os
model = LiteLLMModel(model_id="claude-sonnet-4-5-20250929", api_key=os.getenv("CLAUDE_API_KEY"))
```

```python
agent = CodeAgent(
    tools=[],
    model=model,
    add_base_tools=True,
    planning_interval=3,
    max_steps=5,
    additional_authorized_imports=["pandas", "csv", "open", "os"]
)

# Now the agent can search the web!
result = agent.run(full_prompt)
print(result)
```
